**Importing all the dependencies**


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data collection and preprocessing**

```python
raw_data = pd.read_csv('/content/mail_data.csv')
```

```python
#lets check the data
print(raw_data)
```

```python
#replacing the null values with a null string
mail_data = raw_data.where((pd.notnull(raw_data)),'')
```

```python
# printing the first five rows of the data
mail_data.head()
```

Labeling the data:
spam mail as 1
ham mail as 0

```python
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 1
```

```python
mail_data.head()
```

```python
#lets separating the data as text and label
x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
print(y)
```

Splitting the data into training data & test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
# transform the text data to feature vectors that can be used as input to the logistics regression
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

#coverting the data into integers from  sample set of data

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

Training the model

```python
# training the logistics Regression model with the training data
model = LogisticRegression()
```

```python
model.fit(x_train_features, y_train)
```

Evaluating the trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print("Accuracy on training data: ", accuracy_on_training_data)
```

```python
# lets check prediction on test data

prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
print("Accuracy on test data: ", accuracy_on_test_data)
```

Building a predictive System

```python
input_mail = ["I'm gonna be home soon and i don't want to talk about this stuff anymore tonight, k? I've cried enough today."]

#convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

#making prediction

prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('spam mail')

else:
  print('Ham mail')
```
